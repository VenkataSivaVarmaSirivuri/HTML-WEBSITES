```python
def grouper(df,idx,col):
    if 80<=df[col].loc[idx]<=90:
        return '80-90'
    elif 50<=df[col].loc[idx]<=60:
        return '50-60'
    else:
        return 'others'
```

```python
g=df.groupby(lambda idx: grouper(df,idx,'temperature'))
g
```

```python
for key,d in g:
    print(key)
    print(d)
```

```python
import pandas as pd
df1=pd.DataFrame({'city':['new york','chicago','orlando','baltimore'],
                 'temperature':[21,14,35,38]})
df1
```

```python
df2=pd.DataFrame({'city':['chicago','new york','san diego'],
                 'humidty':[65,68,75]})
df2
```

# i want to combine or merge this

```python
pd.merge(df1,df2,on='city',how='inner')
```


# by default also it is inner

```python
pd.merge(df1,df2,on='city')
```

# left join

```python
pd.merge(df1,df2,on='city',how='left')# df1 left table whatever there in df1 will print with same on right
```

```python
pd.merge(df1,df2,on='city',how='right')#whatever there on right
# will print in left aslo right one only will print
```

```python
pd.merge(df1,df2,on='city',how='outer')# it will print everything
```
